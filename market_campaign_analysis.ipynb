A good approach might be to fill the missing income values as the average of the corresponding education level as opposed to a simple general average

```python
df1 = df.copy()
```

```python
def fill_nas(df):
    for i in df['Education'].unique():
        df[df['Education'] == i] = df[df['Education'] == i].fillna(df[df['Education'] == i]['Income'].mean())
```

```python
fill_nas(df1)
```

```python
expenditure_cols = ['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']
```

```python
def expenditure_scatter_graphs(df,column_name):
    fig, axes = plt.subplots(int(len(expenditure_cols)/2),2,figsize=(20,20))
    for index,col in enumerate(expenditure_cols):
        plt.subplot(int(len(expenditure_cols)/2),2,index+1)
        sns.scatterplot(data=df,x=column_name,y=col)
        plt.tight_layout()
```

```python
expenditure_scatter_graphs(df1,'Year_Birth')
```

```python
plt.figure(figsize=(20,5))
sns.catplot(data=df1,x='Education',col='Response',kind='count')
```
