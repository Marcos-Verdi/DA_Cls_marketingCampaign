A good approach might be to fill the missing income values as the average of the corresponding education level as opposed to a simple general average

```python
df1 = df.copy()
```

```python
def fill_nas(df):
    for i in df['Education'].unique():
        df[df['Education'] == i] = df[df['Education'] == i].fillna(df[df['Education'] == i]['Income'].mean())
```

```python
fill_nas(df1)
```
